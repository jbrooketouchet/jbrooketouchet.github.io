# Part 3

Includes data on the map from the dataframe with differentiated marks

```python
# determines the categories in Crime Name1
df['Crime Name1'].unique()
```

```python
# determines the categories in Crimes Agaisnt Persons
df[df['Crime Name1'] == 'Crime Against Person']['Crime Name2'].unique()
```

```python
# puts the markers on the map

for index, row in df[df['Crime Name1'] == 'Crime Against Person'].iterrows():
    
    if row["Crime Name2"] == "Justifiable Homicide":
        folium.Marker(location=[row["Latitude"], row["Longitude"]],
                    icon=folium.Icon(color='green')).add_to(map_crime_osm)
        
    if row["Crime Name2"] == "Murder and Nonnegligent Manslaughter":
        folium.Marker(location=[row["Latitude"], row["Longitude"]],
                    icon=folium.Icon(color='blue')).add_to(map_crime_osm)
        
    if row["Crime Name2"] == "Negligent Manslaughter":
        folium.Marker(location=[row["Latitude"], row["Longitude"]],
                    icon=folium.Icon(color='red')).add_to(map_crime_osm)

print('Homicide, Murder, and Manslaughter in Montgomery County Maryland')
print('Green: Justifiable Homicide')
print('Blue: Murder and Nonnegligent Manslaughter')
print('Red: Negligent Manslaughter')
map_crime_osm
```

There was a lot of data so I needed to exclude some, so I decided to only focus on crimes agaisnt persons. Crimes agaist persons still included a lot of data so I needed to still exclude some more. I decided to only focus on crimes that resulted in death (homicide, murder, and manslaughter). There were 3 variables corresponding to this: justifiable Homicide, Murder and Nonnegligent Manslaughter, and Negligent Manslaughter. I made them all different colors so you could differetiate betweeen them. I made the Negligent Manslaughter red since there was only one so I wanted it to stand out.
